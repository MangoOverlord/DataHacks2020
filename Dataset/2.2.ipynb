```python
#avg price
data3 = pd.DataFrame()
data3["Census Tract Name"] = data["Census Tract Name"]
data3["Block Group"] = data["Block Group"]
data3["Housing Units"] = data["Total housing units"]
data3["Average Price"] = data.iloc[:,105]

data3["Average Price"] = data3["Average Price"].str.replace("$","")
data3["Average Price"] = data3["Average Price"].str.replace(",","")
data3["Average Price"] = data3["Average Price"].str.replace(".00","")
data3["Average Price"] = data3["Average Price"].replace("...", np.nan)
data3["Average Price"] = data3["Average Price"].replace(" ... ", np.nan)
data3["Average Price"] = data3["Average Price"].replace(" -   ", np.nan)
data3 = data3.dropna()
data3["Average Price"] = data3["Average Price"].astype(str).astype(int)
data3["Housing Units"] = data3["Housing Units"].str.replace(",","")
data3["Housing Units"] = data3["Housing Units"].astype(str).astype(int)

data3["Average Price"].max

housingsum = 0
total = 0
for index, row in data3.iterrows():
    total += row["Average Price"] * row["Housing Units"]
    housingsum += row["Housing Units"]
    
total / housingsum
```
