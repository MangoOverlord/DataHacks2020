```python
temp = populationdata.loc[:,'Male persons age under 5 years':'Male persons age 75 and over']

temp = temp.replace("...", np.nan)
temp = temp.dropna()

for col in temp:
    temp[col] = temp[col].astype(str).astype(int)


temp2 = populationdata.loc[:,'Female persons age under 5 years':'Female persons age 75 and over']

temp2 = temp2.replace("...", np.nan)
temp2 = temp2.dropna()

for col in temp2:
    temp2[col] = temp2[col].astype(str).astype(int)

result = pd.DataFrame()
result["total male"] = temp.sum(axis=1)
result["total female"] = temp2.sum(axis=1)
result["total persons"] = populationdata["Total persons"]
result
```

```python
for col in populationdata.loc[:,'Male persons age under 5 years':'Female persons age 75 and over']:
    populationdata[col] = populationdata[col].replace("...", np.nan)

populationdata = populationdata.dropna()

for col in populationdata.loc[:,'Male persons age under 5 years':'Female persons age 75 and over']:
    populationdata[col] = populationdata[col].astype(str).astype(int)


populationdata["Total Male"] = populationdata.loc[:,'Male persons age under 5 years':'Male persons age 75 and over'].sum(axis=1)
populationdata["Total Female"] = populationdata.loc[:,'Female persons age under 5 years':'Female persons age 75 and over'].sum(axis=1)
populationdata
```
